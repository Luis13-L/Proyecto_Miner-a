### Realizamos un filtro para poder visualizar la severidad en cada caso

```python
Gravedad_columnas = data.filter(like='Gravedad').columns
```

```python
data['Gravedad-Ninguna'].replace({1:'Ninguna',0:'No'},inplace =True)
data['Gravedad-leve'].replace({1:'leve',0:'No'},inplace =True)
data['Gravedad-Moderada'].replace({1:'Moderada',0:'No'},inplace =True)
data['Gravedad_Severa'].replace({1:'Severa',0:'No'},inplace =True)
```

Creamos columna en donde mostraremos la condición 

```python
data['Condicion']=data[Gravedad_columnas].values.tolist()
```

```python
data.head()
```

```python
def remove(lista):
    lista = set(lista) 
    lista.discard("No")
    final = ''.join(lista)
    return final
```

```python
data['Condicion'] = data['Condicion'].apply(remove)
```
